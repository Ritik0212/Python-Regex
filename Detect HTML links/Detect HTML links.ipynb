```python
import re

n= int(input())

while (n):
    k = input()
    k = re.findall(r'<a.*?href="(.*?)".*?>([\w\s.,\/]*)<\/\w>',k)
    for i in k:
        print(','.join([t.strip() for t in i]))
    n = n - 1
    #<a href="(.*?)".*?>([\w .,\/]*)(?=<\/),
```
